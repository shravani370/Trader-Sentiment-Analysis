## Key Insights

### Insight 1 — Profitability varies across sentiment regimes
Trader performance shows clear variation between Fear and Greed market conditions.  
Average realized PnL and win probability tend to improve during Greed periods, while Fear regimes are associated with relatively weaker outcomes.

This indicates that positive sentiment phases may provide more favourable trading opportunities, whereas Fear-driven markets introduce higher uncertainty and inconsistent performance.

### Insight 2 — Traders increase risk exposure during Greed sentiment
Trading behaviour becomes more aggressive during Greed periods.  
Traders execute more trades and take larger position sizes, reflecting higher participation and risk tolerance in bullish market environments.

This behavioural shift suggests increased confidence and a tendency toward momentum-based trading strategies.

### Insight 3 — Trading style influences sentiment sensitivity
Segment analysis shows that trader behaviour significantly impacts performance outcomes.

- Frequent traders demonstrate higher profitability potential but also experience greater variability in returns.
- Infrequent traders display relatively stable performance with lower sensitivity to sentiment changes.
- Consistent traders maintain stronger win rates across both Fear and Greed regimes, highlighting the role of disciplined execution and effective risk management.

### Insight 4 — Overtrading can reduce performance efficiency
Very high trading frequency during Greed sentiment does not always result in proportionally higher profitability.  
Beyond a certain level of participation, transaction costs and execution inefficiencies may reduce overall performance.

This suggests the presence of overconfidence and suboptimal trade timing among highly active traders.

---

## Strategy Recommendations

- **Risk Moderation in Fear Regimes:** Traders should consider reducing trade frequency and position size during Fear periods to manage downside risk and preserve capital.

- **Controlled Participation in Greed Regimes:** While Greed periods offer better opportunities, maintaining disciplined exposure and avoiding excessive leverage can help prevent performance deterioration due to overtrading.

- **Behaviour-Based Strategy Optimization:** Trading approaches can be improved by aligning execution intensity with trader profiles.  
Frequent traders may dynamically adjust participation based on sentiment cycles, while conservative traders may benefit from maintaining stable and risk-adjusted exposure.

---

## Conclusion

This analysis highlights that market sentiment plays an important role in shaping trader behaviour and performance outcomes.  
Traders tend to take greater risks and trade more actively during Greed periods, whereas Fear regimes are associated with lower profitability and higher uncertainty.  
The findings also suggest that disciplined and consistent traders are better positioned to maintain stable performance across different sentiment conditions.  
Overall, incorporating sentiment awareness and behaviour-based adjustments can support more effective trading decisions and risk management.
